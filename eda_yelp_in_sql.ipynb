```python
import pandas as pd
import sqlite3

import warnings
warnings.filterwarnings('ignore')
```

```python
conn = sqlite3.connect('yelper.db')
```

```python
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type = 'table'", conn)
tables
```

```python
for table in tables['name']:
    print('-' * 50, f'{table}', '-' * 50)
    count_df = pd.read_sql(f"SELECT COUNT(*) as count FROM {table}", conn)
    print('Count of records:', count_df['count'].values[0])
    display (pd.read_sql(f"select * from {table} limit 5", conn))
```

```python
business_id = pd.read_sql_query("""select * from yelp_academic_dataset_business 
where lower(categories) like '%restaurant%' and is_open = 1""", conn)
business_id.shape
```

```python
ratings = pd.read_sql_query (f"""select 
avg(review_count) as Avg_reveiw_cnt, min(review_count) as min_review_cnt, max(review_count) as max_review_cnt,

(select review_count from yelp_academic_dataset_business order by review_count limit 1 offset 
(select count(*) from yelp_academic_dataset_business)/2) as median_review_count,

avg(stars) as Avg_rating, min(stars) as min_rating, max(stars) as max_rating,
(select stars from yelp_academic_dataset_business order by stars limit 1 offset 
(select count(*) from yelp_academic_dataset_business)/2) as median_rating

from yelp_academic_dataset_business where business_id in {tuple(business_id['business_id'])}""", conn).transpose()

ratings 
```

```python
# to remove outliers
def remove_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
business_id =  remove_outliers(business_id, 'review_count')
business_id.shape
```

```python
ratings = pd.read_sql_query (f"""select 
avg(review_count) as Avg_reveiw_cnt, min(review_count) as min_review_cnt, max(review_count) as max_review_cnt,

(select review_count from yelp_academic_dataset_business where business_id in {tuple(business_id['business_id'])} order by review_count limit 1 offset 
(select count(*) from yelp_academic_dataset_business where business_id in {tuple(business_id['business_id'])})/2) as median_review_count,

avg(stars) as Avg_rating, min(stars) as min_rating, max(stars) as max_rating,

(select stars from yelp_academic_dataset_business where business_id in {tuple(business_id['business_id'])} order by stars limit 1 offset 
(select count(*) from yelp_academic_dataset_business where business_id in {tuple(business_id['business_id'])})/2) as median_rating

from yelp_academic_dataset_business where business_id in {tuple(business_id['business_id'])}""", conn).transpose()

ratings
```
